### Visualization for Weeknd Forecast (Prophet)

```python
# Load dataset
artist_data = pd.read_csv('Dataset/Final_data.csv')  # Replace with your dataset path

# Aggregate data by Artist and Date
time_series_data = artist_data.groupby(['Date', 'artist'])['points'].sum().reset_index()

# Convert the 'Date' column to datetime format
time_series_data['Date'] = pd.to_datetime(time_series_data['Date'])

# Choose a single artist
artist_name = 'The Weeknd'  # Replace with the desired artist's name
artist_data_filtered = time_series_data[time_series_data['artist'] == artist_name]

# Prepare the data for Prophet
prophet_data = artist_data_filtered[['Date', 'points']].rename(columns={'Date': 'ds', 'points': 'y'})

# Train the Prophet model
model = Prophet()
model.fit(prophet_data)

# Create future dataframe for predictions
future_dates = model.make_future_dataframe(periods=365, freq='D')  # Predicting for 1 year (2024)
forecast = model.predict(future_dates)

# Extract confidence intervals
lower_bound = forecast['yhat_lower']
upper_bound = forecast['yhat_upper']

# Plot Prophet forecast with customized visuals
plt.figure(figsize=(14, 7))

# Historical data
plt.plot(prophet_data['ds'], prophet_data['y'], label='Historical Data', color='black', linewidth=2)

# Forecast line
plt.plot(forecast['ds'], forecast['yhat'], label='Forecast 2024', color='blue', linewidth=2)

# Confidence intervals
plt.fill_between(forecast['ds'], lower_bound, upper_bound, color='blue', alpha=0.2, label='Confidence Interval (±1 STD)')

# Title and axis labels
plt.title(f'Forecast for {artist_name} (2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Points', fontsize=12)

# Grid and legend
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)

# Layout adjustments
plt.tight_layout()
plt.show()
```

### Visualization for Weeknd Forecast (LightGBM)

```python
# Load dataset
file_path = 'data/Final_data.csv'
artist_data = pd.read_csv(file_path)

# Data Preparation
artist_data['Date'] = pd.to_datetime(artist_data['Date'])
time_series_data = artist_data.groupby(['Date', 'artist'])['points'].sum().reset_index()

# Fill missing dates for each artist
artists = time_series_data['artist'].unique()
filled_data = []

for artist in artists:
    artist_data_filtered = time_series_data[time_series_data['artist'] == artist]
    artist_data_full = artist_data_filtered.set_index('Date').asfreq('D', fill_value=0).reset_index()
    artist_data_full['artist'] = artist
    filled_data.append(artist_data_full)

time_series_data_full = pd.concat(filled_data)

# Feature Engineering
time_series_data_full['day'] = time_series_data_full['Date'].dt.day
time_series_data_full['month'] = time_series_data_full['Date'].dt.month
time_series_data_full['year'] = time_series_data_full['Date'].dt.year

# Filter data for a specific artist
artist_name = "The Weeknd"
artist_data_filtered = time_series_data_full[time_series_data_full['artist'] == artist_name]

# Train the LightGBM model
model_params = {
    'n_estimators': 100,
    'learning_rate': 0.1,
    'max_depth': -1,
    'num_leaves': 31,
    'min_child_samples': 20,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': 42
}
model = LGBMRegressor(**model_params)
X = artist_data_filtered[['day', 'month', 'year']]
y = artist_data_filtered['points']
model.fit(X, y)

# Generate predictions for 2024
future_dates = pd.date_range(start='2017-01-01', end='2024-12-31')
future_features = pd.DataFrame({
    'day': future_dates.day,
    'month': future_dates.month,
    'year': future_dates.year
})
predicted_points = model.predict(future_features)

# Combine historical and predicted data for visualization
historical_data = artist_data_filtered.set_index('Date')[['points']]
forecast_data = pd.DataFrame({'Date': future_dates, 'points': predicted_points}).set_index('Date')

# Plot historical and forecast data
plt.figure(figsize=(14, 7))
plt.plot(historical_data.index, historical_data['points'], label='Historical Data', color='black')
plt.plot(forecast_data.index, forecast_data['points'], label='Forecast 2024', color='blue')
plt.fill_between(
    forecast_data.index,
    forecast_data['points'] - np.std(predicted_points),
    forecast_data['points'] + np.std(predicted_points),
    color='blue',
    alpha=0.2,
    label='Confidence Interval (±1 STD)'
)
plt.title(f'Forecast for {artist_name} (2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Points', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
