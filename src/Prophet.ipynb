```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
artist_data = pd.read_csv('Data/Artist_Pop_App.csv')
```

```python
# Aggregate data by Artist, Year, and Month
time_series_data = artist_data.groupby(['Year', 'Month', 'Artist (Ind.)'])['Appearances'].sum().reset_index()

# Ensure a consistent time series for each artist by filling missing months
artists = time_series_data['Artist (Ind.)'].unique()
filled_data = []
```

```python
for artist in artists:
    artist_data_filtered = time_series_data[time_series_data['Artist (Ind.)'] == artist]
    artist_data_filtered['Date'] = pd.to_datetime(artist_data_filtered[['Year', 'Month']].assign(Day=1))
    artist_data_full = artist_data_filtered.set_index('Date').asfreq('MS', fill_value=0).reset_index()
    artist_data_full['Artist (Ind.)'] = artist
    filled_data.append(artist_data_full)

time_series_data_full = pd.concat(filled_data)
```

#### Fiting in the Model

```python
# Time series prediction for each artist
forecast_results = []

for artist in artists:
    # Filter data for the artist
    artist_data_filtered = time_series_data_full[time_series_data_full['Artist (Ind.)'] == artist]
    prophet_data = artist_data_filtered[['Date', 'Appearances']].rename(columns={'Date': 'ds', 'Appearances': 'y'})

    # Check if enough data is available for Prophet
    if prophet_data.shape[0] < 10:
        print(f"Skipping artist '{artist}' due to insufficient data for Prophet.")
        continue  # Skip to the next artist if not enough data

    # Train the Prophet model
    model = Prophet()
    model.fit(prophet_data)

    # Create future dataframe for predictions
    future_dates = model.make_future_dataframe(periods=15, freq='M')  # Predicting till the end of 2024
    forecast = model.predict(future_dates)

    # Extract predictions for 2024
    forecast_2024 = forecast[forecast['ds'].dt.year == 2024]
    forecast_results.append({
        'Artist': artist,
        'Total Points (2024)': forecast_2024['yhat'].sum()
    })
```

```python
# Prepare the dataset for analysis
artist_data['Date'] = pd.to_datetime(artist_data[['Year', 'Month']].assign(Day=1))
time_series_data = artist_data.groupby(['Date', 'Artist (Ind.)'])['Appearances'].sum().reset_index()

# Select the artist for analysis
artist = 'Drake'  # Replace with desired artist
artist_df = time_series_data[time_series_data['Artist (Ind.)'] == artist]

# Skip if no data for the artist
if artist_df.empty:
    print(f"No data available for artist {artist}.")
else:
    # Prepare the data for Prophet
    prophet_data = artist_df[['Date', 'Appearances']].rename(columns={'Date': 'ds', 'Appearances': 'y'})

    # Train-Test Split
    train_data = prophet_data[prophet_data['ds'] < '2023-01-01']
    test_data = prophet_data[prophet_data['ds'] >= '2023-01-01']

    # Skip evaluation if not enough test data
    if test_data.empty:
        print("Not enough test data for evaluation.")
    else:
        # Train the Prophet model
        model = Prophet()
        model.fit(train_data)

        # Generate future dates for the test period
        future_dates = model.make_future_dataframe(periods=len(test_data), freq='MS')  # Monthly start
        forecast = model.predict(future_dates)

        # Filter the forecast for the test period
        forecast_test = forecast[forecast['ds'].isin(test_data['ds'])]

        # Check for length mismatch
        if len(forecast_test) != len(test_data):
            print("Skipping evaluation: inconsistent prediction length.")
        else:
            predictions = forecast_test['yhat'].values  # Predicted values
            actuals = test_data['y'].values  # Actual values

            # Calculate Metrics
            mae = mean_absolute_error(actuals, predictions)
            rmse = np.sqrt(mean_squared_error(actuals, predictions))

            # Display Metrics
            metrics_table = pd.DataFrame({
                'Metric': ['Mean Absolute Error (MAE)', 'Root Mean Squared Error (RMSE)'],
                'Value': [mae, rmse]
            })

            print("Prophet Model Evaluation Metrics")
            print(metrics_table)
```

```python
# Combine all results into a DataFrame
forecast_summary = pd.DataFrame(forecast_results)

forecast_summary

#forecast_summary.to_csv('Data/prophet_App.csv')
```

### Prophet using Daily Data

```python
# Load dataset
artist_data = pd.read_csv('Data/Final_data.csv')  # Replace with your dataset path

# Aggregate data by Artist and Date
time_series_data = artist_data.groupby(['Date', 'artist'])['points'].sum().reset_index()

# Convert the 'Date' column to datetime format
time_series_data['Date'] = pd.to_datetime(time_series_data['Date'])

# Ensure a consistent time series for each artist by filling missing dates
artists = time_series_data['artist'].unique()
filled_data = []
```

```python
for artist in artists:
    artist_data_filtered = time_series_data[time_series_data['artist'] == artist]
    artist_data_full = artist_data_filtered.set_index('Date').asfreq('D', fill_value=0).reset_index()
    artist_data_full['artist'] = artist
    filled_data.append(artist_data_full)

time_series_data_full = pd.concat(filled_data)
```

#### Fiting in the Model

```python
# Time series prediction for each artist
forecast_results = []

for artist in artists:
    # Filter data for the artist
    artist_data_filtered = time_series_data_full[time_series_data_full['artist'] == artist]
    prophet_data = artist_data_filtered[['Date', 'points']].rename(columns={'Date': 'ds', 'points': 'y'})
    
    # Skip artists with less than 12 data points
    if len(prophet_data) < 12:
        forecast_results.append({
            'Artist': artist,
            'Total Points 2024': 0  # Default to 0 if insufficient data
        })
        continue

    # Train the Prophet model
    model = Prophet()
    model.fit(prophet_data)

    # Create future dataframe for predictions
    future_dates = model.make_future_dataframe(periods=365, freq='D')  # Predicting for 1 year (2024)
    forecast = model.predict(future_dates)

    # Extract predictions for 2024
    forecast_2024 = forecast[(forecast['ds'] >= '2024-01-01') & (forecast['ds'] <= '2024-12-31')]
    forecast_results.append({
        'Artist': artist,
        'Total Points 2024': forecast_2024['yhat'].sum()
    })
```

```python
#Evaluation
# Load dataset
artist_data = pd.read_csv('Dataset/final_data.csv')  # Replace with your dataset path

# Aggregate data by Artist and Date
time_series_data = artist_data.groupby(['Date', 'artist'])['points'].sum().reset_index()

# Convert the 'Date' column to datetime format
time_series_data['Date'] = pd.to_datetime(time_series_data['Date'])

# Choose a single artist
artist_name = 'Drake'  # Replace with the desired artist's name
artist_data_filtered = time_series_data[time_series_data['artist'] == artist_name]


# Handle Outliers in the 'points' Column
q1 = artist_data_filtered['points'].quantile(0.25)
q3 = artist_data_filtered['points'].quantile(0.75)
iqr = q3 - q1

# Define outlier bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Replace outliers with bounds
artist_data_filtered['points'] = artist_data_filtered['points'].clip(lower=lower_bound, upper=upper_bound)


# Prepare the data for Prophet
prophet_data = artist_data_filtered[['Date', 'points']].rename(columns={'Date': 'ds', 'points': 'y'})

# Split the data into training and testing sets
train_data = prophet_data[prophet_data['ds'] < '2023-01-01']
test_data = prophet_data[prophet_data['ds'] >= '2023-01-01']

# Train the Prophet model on the training data
model = Prophet()
model.fit(train_data)

# Make predictions for the testing period
future_dates = model.make_future_dataframe(periods=len(test_data), freq='D')  # Only for the test period
forecast = model.predict(future_dates)

# Extract predictions for the testing period
forecast_test = forecast[forecast['ds'].isin(test_data['ds'])]
predictions = forecast_test['yhat'].values  # Predicted values
actuals = test_data['y'].values  # Actual values

# Evaluate the model
mae = mean_absolute_error(actuals, predictions)
mse = mean_squared_error(actuals, predictions)
rmse = np.sqrt(mse)
# Display the metrics
metrics_table = pd.DataFrame({
    'Metric': ['Mean Absolute Error (MAE)', 'Root Mean Squared Error (RMSE)'],
    'Value': [mae, rmse]
})

print(metrics_table)
```
