```python
import pandas as pd
```

```python
df = pd.read_csv('Data/Artist_pop.csv')
```

```python
df = df.drop('Unnamed: 0',axis = 1)
```

```python
df.head()
```

```python
if 'Artist (Ind.)' in df.columns:
    columns = ['Artist (Ind.)'] + [col for col in df.columns if col != 'Artist (Ind.)']
    df = df[columns]
```

```python
df = df.drop(columns=['Title', 'Artists', 'Instrumentalness', '# of Artist', '# of Nationality', 'Nationality', 'Points (Ind for each Artist/Nat)'])
```

```python
# Define the list of continents mapped to numbers
continent_map = {
    'Africa': 0,
    'Asia': 1,
    'Oceania': 2,
    'Europe': 3,
    'Latin-America': 4,
    'Anglo-America': 5
}

# Replace continent names with their corresponding numbers
df['Continent'] = df['Continent'].map(continent_map)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date').reset_index(drop=True)
```

```python
# Extract year and month to use for grouping
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
# Define the columns to average
columns_to_average = ['Rank', 'Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Valence', 'Points (Total)']
```

```python
# Group by 'Artist', 'Year', and 'Month', calculate the mean for specified columns and count occurrences
df_monthly_summary = df.groupby(['Artist (Ind.)', 'Year', 'Month']).agg(
    {**{col: 'mean' for col in columns_to_average}, 'Artist (Ind.)': 'count'}
).rename(columns={'Artist (Ind.)': 'Appearances'}).reset_index()
```

```python
# Sort by year and month
df_monthly_summary = df_monthly_summary.sort_values(by=['Year', 'Month']).reset_index(drop=True)
```
