```python
import pandas as pd
```

```python
df = pd.read_csv('Data/Artist_pop.csv')
```

```python
df = df.drop('Unnamed: 0',axis = 1)
```

```python
df['is_feat'] = df['Artists'].apply(lambda x: 1 if ',' in str(x) or '&' in str(x) else 0)
```

```python
# Drop the specified columns
df.drop(['Title', 'Artists', 'Instrumentalness', '# of Artist', '# of Nationality', 
         'Nationality', 'Points (Total)'], axis=1, inplace=True)
```

```python
df.head()
```

```python
continent_map = {
    'Unknown' : 0,
    'Africa': 1,
    'Asia': 2,
    'Oceania': 3,
    'Europe': 4,
    'Latin-America': 5,
    'Anglo-America': 6
}

# Replace continent names with their corresponding numbers
df['Continent'] = df['Continent'].map(continent_map)
```

```python
continent_counts = df['Continent'].value_counts()
continent_counts
```

```python
# Rename the columns
df.rename(columns={'Artist (Ind.)': 'artist', 'Points (Ind for each Artist/Nat)': 'points'}, inplace=True)

# Move 'artist' to the beginning and 'points' to the end
columns_order = ['artist'] + [col for col in df.columns if col not in ['artist', 'points']] + ['points']
df = df[columns_order]
```

```python
df.head()
```

```python
# Convert the 'date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract the year from the 'date' column
df['year'] = df['Date'].dt.year

# Group by artist and year to calculate the count
artist_year_counts = df.groupby(['artist', 'year']).size().reset_index(name='appearances')
```

```python
df = df.sort_values(by='Date')
```

```python
df[(df['artist'] == 'Drake') & (df['Date'] == '2018-01-01')]
```

```python
# Identify artists that appear only in 2023
artists_in_2023 = df[df['year'] == 2023]['artist'].unique()
artists_in_other_years = df[df['year'] != 2023]['artist'].unique()

# Find the difference to get artists exclusive to 2023
artists_only_in_2023 = set(artists_in_2023) - set(artists_in_other_years)

# Remove rows with artists exclusive to 2023
filtered_data = df[~df['artist'].isin(artists_only_in_2023)]
```

```python
filtered_data
```

```python
aggregated_data = filtered_data.groupby(['Date', 'artist'], as_index=False).agg({
    'Danceability': 'mean',  # Replace with 'sum' if summing makes sense
    'Energy': 'mean',        # Replace with relevant aggregation method
    'Loudness': 'mean',
    'Speechiness': 'mean',
    'Acousticness': 'mean',
    'Valence': 'mean',
    'is_feat': 'mean',
    'Continent': 'first',    # Use 'first' if Continent is categorical       # Sum if you want to count features
    'points': 'sum',         # Sum makes sense for popularity points     # Sum total appearances
})
```

```python
artist_counts = aggregated_data['artist'].value_counts()
Final_data = aggregated_data[aggregated_data['artist'].isin(artist_counts[artist_counts >= 24].index)]
```

```python
Final_data
```

```python
least_present_artist = Final_data['artist'].value_counts().idxmin()
least_present_count = Final_data['artist'].value_counts().min()

print(f"Artist present the least number of times: {least_present_artist}")
print(f"Number of occurrences: {least_present_count}")
```
